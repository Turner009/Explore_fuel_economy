```python
import pandas as pd

df = pd.read_csv("vehicles.csv")
df.info()

```

Build custom database for analyst

```python
explore_fuel_economy = df[['year', 'make', 'model', 'VClass','cylinders', 'displ', 'highway08', 'comb08',
                          'pv2', 'lv2', 'pv4', 'lv4', 'hpv', 'hlv', 'drive']].copy()
explore_fuel_economy = explore_fuel_economy.rename(columns={'comb08':'CombCityHwy MPG',
                                                        'hlv':'HatchLuggVol', 'lv2':'2D LuggVol', 'lv4':'4D LuggVol', 
                                                        'pv2': '2D PassVol', 'highway08': 'HW mpg@55', 'pv4':'4D PassVol',
                                                          'hpv':'HatchPassVol'})
explore_fuel_economy['HW mpg@65'] = (explore_fuel_economy['HW mpg@55']*.9).astype(int)
explore_fuel_economy['HW mpg@75'] = (explore_fuel_economy['HW mpg@65']*.89).astype(int)
explore_fuel_economy['HW mpg@85'] = (explore_fuel_economy['HW mpg@75']*.88).astype(int)
explore_fuel_economy['HW mpg@95'] = (explore_fuel_economy['HW mpg@85']*.87).astype(int)
explore_fuel_economy['HW mpg@105'] = (explore_fuel_economy['HW mpg@95']*.85).astype(int)
explore_fuel_economy = explore_fuel_economy[['year', 'make', 'model', 'VClass', 'cylinders', 'displ', 
                          'drive','CombCityHwy MPG', 'HW mpg@55','HW mpg@65',
                          'HW mpg@75', 'HW mpg@85', 'HW mpg@95', 'HW mpg@105']]
explore_fuel_economy.head()
```
